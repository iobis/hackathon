Download your target file (search on the GitHub repo: https://github.com/iobis/hackathon/tree/master/notebooks

```python
# Put here the full address for the Quarto file you want to convert
target_file = "https://raw.githubusercontent.com/iobis/hackathon/refs/heads/master/notebooks/R/environmental_data_for_occurrences.qmd"
```

```python
import requests

def download_file(url, file_name):
    response = requests.get(url)

    # Write the response content to a file
    with open(file_name, 'wb') as file:
        file.write(response.content)

    print(f"File downloaded as {file_name}")

# Extract file name from the URL
file_name = target_file.split("/")[-1]

# Download the file
download_file(target_file, file_name)
```

Modify the header to convert to Python notebook with R kernel.

```python
def modify_qmd_file(file_path):
    # Define the text to find and the replacement text
    old_text = """
format:
  html:
    embed-resources: true
"""
    new_text = """
jupyter:
  kernelspec:
    display_name: R
    language: R
    name: ir
format:
  ipynb: default
"""

    # Read the content of the .qmd file
    with open(file_path, 'r') as file:
        content = file.read()

    # Replace the old text with the new text
    updated_content = content.replace(old_text, new_text)

    # Write the updated content back to the file
    with open('r_notebook.qmd', 'w') as file:
        file.write(updated_content)

    print("The .qmd file has been updated successfully.")

# Use the function on your .qmd file
modify_qmd_file(file_name)
```

Render the file.

```python
!quarto render r_notebook.qmd --no-execute
```
